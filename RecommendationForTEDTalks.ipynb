```python
import re
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def remove_tags(string):
    #re.sub for replacement(html ifadeleri geldiyse silmeliyiz)
    result = re.sub('<.*?>','',string)
    return result
```

```python
#details verilerindeki html ifadelerini silmek için
df_txt['details'] = df_txt['details'].apply(lambda cw : remove_tags(cw))
```

```python
#hepsii küçük harf yapıyoruz.
df_txt['details']=df['details'].str.lower()
df_txt['title']=df['title'].str.lower()
```

```python
df_txt
```

```python
desc_text = df_txt['details'].tolist()
tfidf = text.TfidfVectorizer(input = desc_text, stop_words = "english")
desc_matrix = tfidf.fit_transform(desc_text)
print(desc_matrix.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim_desc = cosine_similarity(desc_matrix)
print(cosine_sim_desc)
```

```python
indices = pd.Series(df.index, index = df['title'])
def talk_recommend(url):

    idx = indices[url]
    sim_scores = list(enumerate(cosine_sim_desc[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:31]

    talk_indices = [i[0] for i in sim_scores]

    return df_txt.iloc[talk_indices]
```

```python
#tagleri düzenlemek
import re
```

```python
def clean_text(x):
    letter_only = re.sub("[^a-zA-Z]", " ", x)
    return ' '.join(letter_only.split()).lower()
```

```python
df_txt.tags = df_txt.tags.astype('str')
```

```python
df_txt['tags'] = df_txt['tags'].apply(clean_text)
```

```python
df_txt.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
cv_tags = cv.fit_transform(df_txt['tags'])
#df_genres = pd.DataFrame(cv_tags.todense(), columns = cv.get_feature_names(), index = df_txt['title'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cos_sim = cosine_similarity(cv_tags)
```

```python
def get_recommendation_based_tags(x):
    index_to_search = df[df['title'] == x].index[0]
    series_similar = pd.Series(cos_sim[index_to_search])
    index_similar = series_similar.sort_values(ascending = False).head(10).index
    return df_txt.loc[index_similar]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df=0, stop_words='english')

matrix = tf.fit_transform(df_txt['title'])
```

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_similarities_title = linear_kernel(matrix, matrix)
print(cosine_similarities_title)
```

```python
indices = pd.Series(df.index, index = df['title'])
#print(indices)
def talk_recommend_title(title):

    idx = indices[title]
    sim_scores = list(enumerate(cosine_similarities_title[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:31]

    talk_indices = [i[0] for i in sim_scores]

    return df_txt.iloc[talk_indices]
```

```python
pd.set_option('display.max_colwidth', None)
print("Similars based title : \n")
talk_recommend_title('Do schools kill creativity?').head(10)
```

```python
print("Similars based tags : \n")
get_recommendation_based_tags('Do schools kill creativity?')['url']
```

```python
indices = pd.Series(df.index, index = df['title'])
def talk_recommend_forall(title, tags_rate, title_rate, details_rate):

    idx = indices[title]
    sim_scores_title = list(enumerate(cosine_similarities_title[idx]))
    sim_scores_tags = list(enumerate(cos_sim[idx]))
    sim_scores_desc = list(enumerate(cosine_sim_desc[idx]))
    result = [[0 for x in range(2)] for y in range(df_txt.shape[0])]
    for i in range(0, df_txt.shape[0]):
        result[i][0] = i
        result[i][1] = title_rate * sim_scores_title[i][1] + sim_scores_tags[i][1] * tags_rate + sim_scores_desc[i][1] * details_rate
#oranlar degiskene atılsın, kullanıcı versin
    result = sorted(result, key=lambda x: x[1], reverse=True)

    result = result[1:31]

    talk_indices = [i[0] for i in result]

    return df_txt.iloc[talk_indices]
```

```python
talk_recommend_forall('Do schools kill creativity?', 0.5, 0.1, 0.4).head(10)
```

```python
def printMenu():

    print("\n\n<Welcome to TED Talks Recommendation Project>\n")
    print("Opeations :\n")
    print("1 - Recommendation based title.\n")
    print("2 - Recommendation based tags.\n")
    print("3 - Recommendation based details.\n")
    print("4 - Recommendation based all.\n")
    print("0 - Quit.\n")
```

```python
def rec():
    select = 1
    while(int(select)):
        printMenu()
        select = input("Please select an operation\n\t> ")
        if(int(select) == 1):# title
            title = input("Enter a title:")
            print(talk_recommend_title(title)['url'].head(10))
        elif(int(select) == 2):# tags
            title = input("Enter a title: ")
            print(get_recommendation_based_tags(title)['url'])
        elif(int(select) == 3):# details
            title = input("Enter a title: ")
            print(talk_recommend(title)['url'].head(10))
        elif(int(select) == 4):# all
            tags_rate = input("Please enter the similarity rate of tags: ")
            title_rate = input("Please enter the similarity rate of title: ")
            details_rate = input("Please enter the similarity rate of details: ")
            title = input("Enter a title: ")
            print(talk_recommend_forall(title, float(tags_rate), float(title_rate), float(details_rate)).head(10)['url'])
        elif(int(select) == 0):
            select = 0
        else:
            print("Please enter a valid operation.\n")
```
